```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Read the dataset

df = pd.read_csv("height.csv")
```

```python
df.head()
```

```python
plt.scatter(df["Weight"],df["Height"])
plt.xlabel("Weight")
plt.xlabel("Height")
```

```python
# devide features based on dependent and indipendent feature
X = df[['Weight']] #Dependent feature
y = df[['Height']] #Independent feature
```

```python
#Train test split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state = 42)
```

```python
X.shape
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
#Standardise the data
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
X_train= scalar.fit_transform(X_train)
X_test= scalar.fit_transform(X_test)
```

```python
plt.scatter(X_train,y_train)
```

```python
## train the simple liner regression model
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
print("the slove of weight is",regressor.coef_)
print("intercept",regressor.intercept_)
```

```python
plt.scatter(X_train,y_train)
plt.plot(X_train,regressor.predict(X_train),color="Green")
```

```python
Y_pred_test = regressor.predict(X_test)
```

```python
Y_pred_test,y_test
```

```python
plt.scatter(X_test,y_test)
plt.plot(X_test,regressor.predict(X_test),color="Green")
```

```python
## Performance Matrix

## MSE,MAE,RMAE

## R Squared 

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
mse = mean_squared_error(y_test,Y_pred_test)
mae = mean_absolute_error(y_test,Y_pred_test)
rmse = np.sqrt(mse)

print(mse,mae,rmse)
```

```python
from sklearn.metrics import r2_score

score = r2_score(y_test,Y_pred_test)
score
```
