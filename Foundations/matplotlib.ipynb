```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [10,20,15,30]
plt.plot(x, y)
plt.title("Title")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

```python
names = ["A","B","C"]
marks = [80, 92, 75]

plt.bar(names, marks)
plt.title("Bar Chart")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=100)
plt.title("Histogram")
plt.show()
```

```python
sizes = [40, 30, 20, 10]
labels = ["A", "B", "C", "D"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Pie Chart")
plt.show()
```

```python
fig, ax = plt.subplots(1, 2)

ax[0].plot([1,2,3], [10,20,30])
ax[0].set_title("Plot 1")

ax[1].bar(["A","B","C"], [5,7,9])
ax[1].set_title("Plot 2")

plt.show()


```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, figsize=(8,4))

# left subplot (index 0): line plot
ax[0].plot([2, 3, 4, 5], [5, 2, 7, 2], marker='o', linestyle='-')
ax[0].set_title("Line plot")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")

# right subplot (index 1): bar chart  make sure lengths match
labels = ['a', 'b', 'c']
heights = [8, 8, 4]   # length matches labels
ax[1].bar(labels, heights)
ax[1].set_title("Bar chart")
ax[1].set_xlabel("category")
ax[1].set_ylabel("value")

plt.tight_layout()
plt.show()
```
